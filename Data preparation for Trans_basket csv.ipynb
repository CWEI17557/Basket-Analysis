# Data preparation for Basket Analysis

```python
import pandas as pd
product = pd.read_csv('product.csv')   
product.head()
```

```python
product.nunique()
```

```python
transaction = pd.read_csv('transaction_data.csv')  
trans_valid= transaction.loc[transaction['SALES_VALUE']>0]
trans_prod =pd.merge(trans_valid, product, how = 'left', on = 'PRODUCT_ID')
```

```python
trans_prod.head() 
```

```python
trans_prod.nunique()
```

```python
trans_prod.isnull().sum()
```

```python
trans_basket = trans_prod[["BASKET_ID","COMMODITY_DESC"]]
trans_basket
```

```python
trans_basket.to_csv('trans_basket.csv')
```

```python
trans_basket.groupby('COMMODITY_DESC').size().sort_values(ascending=False)[:10]
```
